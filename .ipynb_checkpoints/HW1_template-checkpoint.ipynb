### 1. Briefly explain how the weighted Erdos-Renyi generative model works.

<span style="color:red">NOTE: Rewrite this part and remember that weighted is asked.
With given node count $n$ and probability $0 \leq p \leq 1$, the Erdos-Renyi model creates an undirected graph. The graph has edges between all pairs of its nodes with probability $p$ for each edge. #Algorithm
In the ER model we first define the N
nodes, and then connect each pair of nodes with probability p. This
algorithm generates a homogeneous network (Fig. 1), whose con-
nectivity follows a Poisson distribution peaked at hki and decaying

### 5. Visualize two random graphs you simulated (ER and SF).

```python
#Creation ER and SF
#ER 
print("Creating Erdös-Renyi graphs...")
erdos_renyi_graphs = []
for c in range(graph_count):
    erdos_renyi_G = nx.erdos_renyi_graph(number_of_nodes, probs_edge_creation[c], directed = False) #undirected
    erdos_renyi_graphs.append(erdos_renyi_G)

for G in erdos_renyi_graphs:
    for (u, v) in G.edges():
        G.edges[u,v]['weight'] = random.randint(min_weight, max_weight) #add weight to ER graphs

#SF
print("Creating Scale-Free graphs...")
scale_free_graphs = []
for c in range(graph_count):
    scale_free_G = nx.scale_free_graph(number_of_nodes) #directed
    scale_free_G = nx.DiGraph.to_undirected(nx.DiGraph(scale_free_G)) #making it undirected
    scale_free_graphs.append(scale_free_G)

for G in scale_free_graphs:
    for (u, v) in G.edges():
        G.edges[u,v]['weight'] = random.randint(min_weight, max_weight) #add weight to SF graphs
```

```python
#Plotting ER and SF
ER_random = random.randint(0, graph_count-1)
SF_random = random.randint(0, graph_count-1)

_, axes = plt.subplots(nrows=1, ncols=2, figsize=(14,7))
axes[0].title.set_text(r"$\mathbf{Sample\ Erdös-Renyi\ Network}$")
axes[1].title.set_text(r"$\mathbf{Sample\ Scale-Free\ Network}$")
nx.draw(erdos_renyi_graphs[ER_random], ax=axes[0], node_color=range(number_of_nodes), node_size=node_draw_size) #plot ER
nx.draw(scale_free_graphs[SF_random], ax=axes[1], node_color=range(number_of_nodes), node_size=node_draw_size) #plot SF
plt.show()
```

### 6. Plot the overlaid distributions of the node strength centrality for all 30 ER graphs (transparent blue color) and the 30 SF graphs (transparent red color). What do you notice? Interpret your observation.

```python
def get_strength(G):
    return dict(G.degree(weight = "weight"))

def calculate_node_strength_centrality(G_list):
    centrality_array = np.empty([graph_count, number_of_nodes])
    i = 0
    for G in G_list:
        strength_values = get_strength(G).values()
        centrality_array[i, :] = np.array(list(strength_values))
        i += 1

    centrality_array = centrality_array.reshape(graph_count * number_of_nodes)
    return centrality_array

er_degree_centrality = calculate_node_strength_centrality(erdos_renyi_graphs)
sf_degree_centrality = calculate_node_strength_centrality(scale_free_graphs)

bins = np.linspace(0, number_of_nodes * max_weight, bin_count) #max possible strength of a node is number_of_nodes * max_weight 
plt.title("Node Strength Centrality")
plt.xlabel("Node strength")
plt.ylabel("Number of nodes")
plt.hist(er_degree_centrality, bins, color = 'b', alpha=.4, label= 'ER')
plt.hist(sf_degree_centrality, bins, color = 'r', alpha=.4, label= 'SF')
plt.legend(loc='upper right')
plt.show()
```

I noticed that while ER graphs' node strength centrality is more like uniformly distributed, SF's is mostly accumulated to values near zero. Because of ER graphs are created by randomly assigning weights to all nodes while SFs have less number of nodes with higher weights, we are expected to see this distribution.

### 7. Plot the overlaid distributions of the eigenvector centrality for all 30 ER graphs (transparent blue color) and the 30 SF graphs (transparent red color). What do you notice? Interpret your observation.

```python
def eigenvector_centrality(G):
    x = nx.eigenvector_centrality(G, weight = "weight", max_iter = max_solver_iterations)
    return x

def calculate_eigenvector_centrality(G_list):
    centrality_array = np.empty([graph_count, number_of_nodes])
    i = 0
    for G in G_list:
        eigenvector_values = eigenvector_centrality(G).values()
        centrality_array[i, :] = np.array(list(eigenvector_values))
        i += 1

    centrality_array = centrality_array.reshape(graph_count * number_of_nodes)
    return centrality_array

er_eigenvector_centrality = calculate_eigenvector_centrality(erdos_renyi_graphs)
sf_eigenvector_centrality = calculate_eigenvector_centrality(scale_free_graphs)

bins = np.linspace(0, 0.5, bin_count) #eigenvector takes values between 0 and 1 
plt.title("Eigenvector Centrality")
plt.xlabel("Eigenvector")
plt.ylabel("Number of nodes")
plt.hist(er_eigenvector_centrality, bins, color = 'b', alpha=.4, label= 'ER')
plt.hist(sf_eigenvector_centrality, bins, color = 'r', alpha=.4, label= 'SF')
plt.legend(loc='upper right')
plt.show()
```

If a node's eigenvector centrality is high, it means the node has many neighbours and/or its neighbours are highly connected. Therefore for SF graph, higher eigenvector centrality values are possible than ERs. As we can see from the histogram, nodes of SF have variable eigenvector centrality values and ERs have similar eigenvector centrality values. It is expected since SF has few highly connected nodes and the nodes closer to them have higher eigenvector centrality values.

### 1. [10p] Code up a function called ‘findPercolationThreshold’, where you input a positive weighted adjacency matrix and it automatically outputs its percolation threshold $f_c$.

```python
def findPercolationThreshold(A):
    G = nx.from_numpy_matrix(A)
    n = G.number_of_nodes()
    e = G.number_of_edges()
    G_new = nx.empty_graph(n=0)
    
    S = []
    connection_density = []
    
    edges = sorted(G.edges(data=True), key=lambda t: t[2].get('weight', 1), reverse=True)
    for edge in edges:
        if not G_new.has_edge(*edge[:2]):
            G_new.add_edge(*edge[:2])
            LCC = G_new.subgraph(max(nx.connected_components(G_new), key=len)).copy() #calculate largest connected component of new graph
            S.append(LCC.number_of_nodes() / n) #node size of the largest connected component normalized with number of nodes of graph G
            connection_density.append(G_new.number_of_edges() / e)
    
    S_array = np.array(S)
    connection_density_array = np.array(connection_density)
    f_c = connection_density[np.argmax(S_array)]
    return f_c, connection_density_array, S_array #percolation threshold
```

### 3. [10p] Code up a function called ‘largestConnectedComponent’, where you input a positive weighted adjacency matrix and it automatically outputs (1) the largest connected component graph, (2) its node size S and (3) its diameter d defined as the average of the shortest paths between all pairs of nodes in the large connected component graph.

```python
def largestConnectedComponent(A): #input is adjacency matrix
    G = nx.from_numpy_matrix(A)
    n = G.number_of_nodes()
    LCC = G.subgraph(max(nx.connected_components(G), key=len)).copy()
    S = LCC.number_of_nodes() / n #LCC's node size S normalized with number of nodes
    d = nx.average_shortest_path_length(LCC) #LCC's diameter d defined as the average of the shortest paths between all pairs of nodes in LCC
    return LCC, S, d
```

### 5. [5p] Random failure scenario. Code up a function called ‘randomError’, inputting (1) the graph adjacency matrix and (2) the fraction (between 0 and 1) of nodes to be randomly removed and returns the adjancy matrix graph excluding the removed nodes. Comment out the code and explain in a text box how it works.

```python
def randomError(A, f):
    G = nx.from_numpy_matrix(A)
    n = G.number_of_nodes()
    node_indexes_to_be_removed = random.sample(G.nodes, int(n * f))
    for i in node_indexes_to_be_removed:
        G.remove_node(i)
    return nx.to_numpy_matrix(G)
```

### 6. [5p] Target attack scenario. Code up a function called ‘targetAttack’, inputting (1) the graph adjacency matrix and (2) the fraction (between 0 and 1) of nodes to be attacked removed and returns the adjancy matrix graph excluding the removed nodes. Note that the nodes to be removed are those with the highest strength. Comment out the code and explain in a text box how it works.

```python
def targetAttack(A, f):
    G = nx.from_numpy_matrix(A)
    n = G.number_of_nodes()
    end_index = int(n * f)
    for i in range(end_index):
        nodes = [node for (node, strength) in sorted(G.degree(weight = "weight"), key=lambda pair: pair[1], reverse=True)]
        node_with_highest_connectivity = nodes[0]
        G.remove_node(node_with_highest_connectivity)
    return nx.to_numpy_matrix(G)
```
