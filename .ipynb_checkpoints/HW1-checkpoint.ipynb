## Part A: Simulate exponential and free-scale networks (35 Points, 5 points each question)

### 1. Briefly explain how the weighted Erdos-Renyi generative model works.
Erdos-Renyi graphs works in a way that all edges between all nodes are created with probability $p$. First the nodes are defined and connection is established between each pair of node with probability $p$. Therefore we obtain a homogoneous network. If the edge is created between nodes, we assign a weight to it for weighted Erdos-Renyi graph.

### 2. What are the key properties of weighted Erdos-Renyi (ER) graphs?

-Undirected graphs with N number of nodes

-Nodes with same degree are equally probable

-The probability of seeing an edge between any two nodes is *p*

### 3. Briefly explain how weighted scale-free (SF) algorithm works.

The algorithm starts by creating $m_0$ nodes. Then each new node is more likely connected to the node with highest links. The probability $\pi_i$ that the new node is connected to node $i$ depends on the connectivity $k_i$ of node $i$ such that $\pi = k_i / \sum_j k_j$.

### 4. What are the key properties of weighted SF graphs?

-Degree distribution follows a power law

-Inhomogeneous

-Less number of nodes have high connectivity

### 5. Visualize two random graphs you simulated (ER and SF).

```python
#Creation ER and SF
#ER 
print("Creating Erdös-Renyi graphs...")
erdos_renyi_graphs = []
for c in range(graph_count): #for each graph
    erdos_renyi_G = nx.erdos_renyi_graph(number_of_nodes, probs_edge_creation[c], directed = False) #create ER graph
    erdos_renyi_graphs.append(erdos_renyi_G) #add created ER graph to list of ER graphs

for G in erdos_renyi_graphs: #for each ER graph
    for (u, v) in G.edges(): #for each edge
        G.edges[u,v]['weight'] = random.randint(min_weight, max_weight) #assign weight to ER graph's edge

#SF
print("Creating Scale-Free graphs...")
scale_free_graphs = []
for c in range(graph_count): #for each graph
    scale_free_G = nx.scale_free_graph(number_of_nodes) # create directed SF graph
    scale_free_G = nx.DiGraph.to_undirected(nx.DiGraph(scale_free_G)) #make SF graph undirected
    scale_free_graphs.append(scale_free_G) #add created SF graph to list of SF graphs

for G in scale_free_graphs: #for each SF graph
    for (u, v) in G.edges(): #for each edge
        G.edges[u,v]['weight'] = random.randint(min_weight, max_weight) #assign weight to SF graph's edge
```

### 6. Plot the overlaid distributions of the node strength centrality for all 30 ER graphs (transparent blue color) and the 30 SF graphs (transparent red color). What do you notice? Interpret your observation.

```python
def get_strength(G):
    return dict(G.degree(weight = "weight")) #get strength (weight of each edges) of each nodes as dictionary

def calculate_node_strength_centrality(G_list):
    centrality_array = np.empty([graph_count, number_of_nodes]) #for each graph, each node has a centrality value
    i = 0
    for G in G_list:
        strength_values = get_strength(G).values() #calculate node strength centrality values
        centrality_array[i, :] = np.array(list(strength_values)) #assign nodes' strength centrality to each graph
        i += 1

    centrality_array = centrality_array.reshape(graph_count * number_of_nodes) # reshape centrality values from 2D to 1D for histogram
    return centrality_array

er_degree_centrality = calculate_node_strength_centrality(erdos_renyi_graphs) # ER node strangth centrality
sf_degree_centrality = calculate_node_strength_centrality(scale_free_graphs) # SF node strangth centrality

bins = np.linspace(0, number_of_nodes * max_weight, bin_count) #max possible strength of a node is number_of_nodes * max_weight 
plt.title("Node Strength Centrality")
plt.xlabel("Node strength")
plt.ylabel("Number of nodes")
plt.hist(er_degree_centrality, bins, color = 'b', alpha=.4, label= 'ER') #ER histogram
plt.hist(sf_degree_centrality, bins, color = 'r', alpha=.4, label= 'SF') #SF histogram
plt.legend(loc='upper right')
plt.show()
```

I noticed that while ER graphs' node strength centrality is more like uniformly distributed, SF's is mostly accumulated to values near zero. Because of ER graphs are created by randomly assigning weights to all nodes while SFs have less number of nodes with higher weights, we are expected to see this distribution.

### 1. [10p] Code up a function called ‘findPercolationThreshold’, where you input a positive weighted adjacency matrix and it automatically outputs its percolation threshold $f_c$.

```python
def findPercolationThreshold(A): #input is adjancency matrix representation of a graph
    G = nx.from_numpy_matrix(A) #convert adjacency matrix to networkx graph
    n = G.number_of_nodes() #number of nodes of the input graph
    e = G.number_of_edges() #number of edges of the input graph
    G_new = nx.empty_graph(n=0) #new graph with no nodes
    
    S = [] #node size list
    connection_density = [] #connection density list
    
    edges = sorted(G.edges(data=True), key=lambda t: t[2].get('weight', 1), reverse=True) #sort edges descending order by weights
    for edge in edges: #for each edge in input graph
        if not G_new.has_edge(*edge[:2]): #if it is not already added to newly created graph G (G_new)
            G_new.add_edge(*edge[:2]) #add edge of input graph G to new edge to G_new (u,v,w)
            LCC = G_new.subgraph(max(nx.connected_components(G_new), key=len)).copy() #get largest connected component of new graph as subgraph
            S.append(LCC.number_of_nodes() / n) #node size of the largest connected component normalized with number of nodes of input graph G
            connection_density.append(G_new.number_of_edges() / e) #connection density as proportion of number of new graph's edges to number of input graph's edges
    
    S_array = np.array(S) #convert node size list to array
    connection_density_array = np.array(connection_density) #convert connection density list to array
    f_c = connection_density[np.argmax(S_array)] #percolation threshold is the value connection density when the node size of LCC hits its heighest value
    return f_c, connection_density_array, S_array #returns percolation threshold, connection density array and node size of LCC array
```

### 2. [5p] Comment out each line of your ‘findPercolationThreshold’ function and add a text box explaining how your algorithm works.

"Percolation efers to processes that occur ona network as nodes and/or edges are added or deleted. Percolation is typically studied in relation to the size of the largest connected component of a graph. A classic example is the rapid emergence of a single giant component that encompasses a large fraction of nodes in an Erdos-Renyi graph at a given probability of connection. The probability at which this large component emerges is called the percolation threshold." [Fundamentals of Brain Network Analysis Book]

To find the percolation threshold I start with an empty graph and add edges of the input graph in descending order of weights. If the edge is already added to new graph I continue with the next edge. At each step I calculate the largest connected component (LCC) of the new graph and save the number of nodes of the LCC to S array. Since generally S is normalized with the number of nodes of the input graph I divided it with n. Since we would like to see the increase in the node size of LCC against connection density, at each step, I calculate the proportion of number of edges of new graph and number of edges of the input graph. At the end for plotting we have S_array for y-axis and connection_density for x-axis. Since at the point where the largest connected component encompasses most of the graph, I take connection density value where the maximum point of number of node size of LCC occurs.

### 3. [10p] Code up a function called ‘largestConnectedComponent’, where you input a positive weighted adjacency matrix and it automatically outputs (1) the largest connected component graph, (2) its node size S and (3) its diameter d defined as the average of the shortest paths between all pairs of nodes in the large connected component graph.

```python
def largestConnectedComponent(A): #input is adjacency matrix
    G = nx.from_numpy_matrix(A) #convert adjacency matrix to networkx graph
    n = G.number_of_nodes() #number of nodes of graph
    LCC = G.subgraph(max(nx.connected_components(G), key=len)).copy() #get largest connected component of graph as subgraph
    S = LCC.number_of_nodes() / n #LCC's node size S is normalized with number of nodes of graph
    d = nx.average_shortest_path_length(LCC) #LCC's diameter d defined as the average of the shortest paths between all pairs of nodes in LCC
    return LCC, S, d #return largest connected component, normalized LCC's node size and LCC's diameter d
```

### 4. [5p] Comment out each line of your ‘largestConnectedComponent’ function and add a text box explaining how your algorithm works.

I first compute the largest connected component (LCC) by using networkx package. Then I calculate of the node size of the LCC and normalize it with the number of nodes of the graph. Therefore maximum S becomes 1. And then I calculate the average of the shortest paths between all pairs of nodes in LCC.

### 5. [5p] Random failure scenario. Code up a function called ‘randomError’, inputting (1) the graph adjacency matrix and (2) the fraction (between 0 and 1) of nodes to be randomly removed and returns the adjancy matrix graph excluding the removed nodes. Comment out the code and explain in a text box how it works.

```python
def randomError(A, f): #adjacency matrix A of the graph and the fraction (f) of nodes to be randomly removed
    G = nx.from_numpy_matrix(A) #convert adjacency matrix to graph
    n = G.number_of_nodes() #get number of nodes of the graph
    node_indexes_to_be_removed = random.sample(G.nodes, int(n * f)) #number of nodes * fraction gives the number of nodes to be removed and i randomly sample this number of node indexes
    for i in node_indexes_to_be_removed: #random indexes of nodes to be removed
        G.remove_node(i) #remove node from graph
    return nx.to_numpy_matrix(G) #return the adjacency matrix of the graph excluding the removed nodes
```

In this function, we define the fraction of nodes to be removed from a graph. I first randomly sample the indexes of the nodes to be removed. Depending on the fraction, the number of nodes to be removed changes. For f=1.0 all nodes to be deleted while f=0.0 there is no node to be deleted. After randomly deleting nodes, new graph is returned.

### 6. [5p] Target attack scenario. Code up a function called ‘targetAttack’, inputting (1) the graph adjacency matrix and (2) the fraction (between 0 and 1) of nodes to be attacked removed and returns the adjancy matrix graph excluding the removed nodes. Note that the nodes to be removed are those with the highest strength. Comment out the code and explain in a text box how it works.

```python
def targetAttack(A, f): #adjacency matrix A of the graph and the fraction (f) of nodes to be randomly removed
    G = nx.from_numpy_matrix(A) #convert adjacency matrix to graph
    n = G.number_of_nodes() #get number of nodes of the graph as n
    end_index = int(n * f) #number of nodes to be removed is calculated by multiplying n with fraction
    for i in range(end_index): #for number of nodes to be removed
        nodes = [node for (node, strength) in sorted(G.degree(weight = "weight"), key=lambda pair: pair[1], reverse=True)] #sort the nodes with node strength in descending order
        node_with_highest_connectivity = nodes[0] #node with highest connectivity is at the index 0 of the sorted list
        G.remove_node(node_with_highest_connectivity) #remove node with the highest connectivity
    return nx.to_numpy_matrix(G) #return the adjacency matrix of the graph excluding the removed nodes
```

For target attack we are attacking the node with the highest connectivity. At the beginning I calculate the number of nodes to be removed considering the fraction. If fraction is 1.0 all nodes will be removed while for 0.0 the graph remains same. For each node to be removed, I first find the node with the highest connectivity. Then I remove it. And continue until all the number of nodes to be removed is reached.

### 7. [5p] Given one random ER network and one random SF network (you select two from your 60 simulated graphs), plot the diameter d of the largest connected component as in Fig. 2 against the fraction of removed nodes in both random error and target attack scenarios. The nodes need to be removed progressively (e.g., remove 0.1, 0.2, until reaching 0.8). Important note: Watch out for the percolation threshold. You might need to remove a very small fraction of nodes before it breaks ( < 0.1 ). Keep an eye on that. Adjust your fraction step accordingly. Produce 3 more similar plots for 6 randomly selected ER and SF graphs (3 ER and 3 SF). Totally, you will have 4 plots for pairs of ER and SF networks to compare.

```python
def getDiameter(sc, G, fraction): #scenario name (sc), graph (G) and fraction array
    i = 0
    d_array = np.zeros(len(fraction)) #diameter array
    
    for f in fraction: #for each fraction starting from f_start and ending at f_end
        if sc == "randomError":
            A = randomError(nx.to_numpy_matrix(G), f) #random failure
        elif sc == "targetAttack":
            A = targetAttack(nx.to_numpy_matrix(G), f) #attack to the node with highest connectivity
        else:
            print("Scenario is invalid!")
            return
        
        if A.shape == (0, 0): #when the all nodes are removed, shape of the adjacency matrix of the graph is (0,0)
            d_array[i] = 0.0
        else: #when the all nodes are not removed
            _, _, d = largestConnectedComponent(A) #compute the diameter of the largest connected component
            d_array[i] = d #put the diameter to diameter array
        i += 1
    return d_array


for i in range(total_number_of_plots): #1 ER and 1 SF are plotted at the same figure and for total_number_of_plots show different sample of graphs
    ER_random = random.randint(0, graph_count-1) #randomly sample ER graph
    SF_random = random.randint(0, graph_count-1) #randomly sample SF graph

    ER = erdos_renyi_graphs[ER_random]
    SF = scale_free_graphs[SF_random]
    
    f_start = 0.0
    f_end = 0.05
    f_sample_n = 100
    fraction = np.linspace(f_start, f_end, num=f_sample_n) #this represents the x axis

    ER_d_array_RE = getDiameter("randomError", ER, fraction) #get the diameter of the LCC for ER failure case
    SF_d_array_RE = getDiameter("randomError", SF, fraction) #get the diameter of the LCC for SF failure case

    ER_d_array_TA = getDiameter("targetAttack", ER, fraction) #get the diameter of the LCC for ER attack case
    SF_d_array_TA = getDiameter("targetAttack", SF, fraction) #get the diameter of the LCC for SF attack case

    plt.figure(figsize=(7, 5))
    plt.title("Failure and Attack Scenarios")
    plt.xlabel("f (fraction of removed nodes)")
    plt.ylabel("d (diameter) of LCC (largest connected component)")
    plt.plot(fraction, ER_d_array_RE, color = 'b', marker='o', label= 'ER Failure') #x-axis is fraction of nodes to be removed, y-axis is diameter of the LCC for ER failure case 
    plt.plot(fraction, SF_d_array_RE, color = 'r', marker='s', label= 'SF Failure') #x-axis is fraction of nodes to be removed, y-axis is diameter of the LCC for SF failure case
    plt.plot(fraction, ER_d_array_TA, color = 'b', marker='>', label= 'ER Attack') #x-axis is fraction of nodes to be removed, y-axis is diameter of the LCC for ER attack case
    plt.plot(fraction, SF_d_array_TA, color = 'r', marker='^', label= 'SF Attack') #x-axis is fraction of nodes to be removed, y-axis is diameter of the LCC for SF attack case
    plt.legend(loc='upper right')
    plt.show()
```

Produce 3 more similar plots for 6 randomly selected ER and SF graphs (3 ER and 3 SF). Totally, you will have 4 plots for pairs of ER and SF networks to compare.

```python
def getS(sc, G, fraction):  #scenario name (sc), graph (G) and fraction array
    i = 0
    S_array = np.zeros(len(fraction)) #array of node size of LCC
    
    for f in fraction:  #for each fraction starting from f_start and ending at f_end
        if sc == "randomError":
            A = randomError(nx.to_numpy_matrix(G), f) #random failure
        elif sc == "targetAttack":
            A = targetAttack(nx.to_numpy_matrix(G), f) #attack to the node with highest connectivity
        else:
            print("Scenario is invalid!")
            return
        
        if A.shape == (0, 0): #when the all nodes are removed, shape of the adjacency matrix of the graph is (0,0)
            S_array[i] = 0.0
        else:
            _, S, _ = largestConnectedComponent(A) #compute the normalized node size (S) of the largest connected component
            S_array[i] = S #put the node size of the LCC to array of node size of LCC
        i += 1
    return S_array


for i in range(total_number_of_plots): #1 ER and 1 SF are plotted at the same figure and for total_number_of_plots show different sample of graphs
    ER_random = random.randint(0, graph_count-1) #randomly sample ER graph
    SF_random = random.randint(0, graph_count-1) #randomly sample SF graph

    ER = erdos_renyi_graphs[ER_random]
    SF = scale_free_graphs[SF_random]
    
    f_start = 0.0
    f_end = 1.0
    f_sample_n = 50
    fraction = np.linspace(f_start, f_end, num=f_sample_n) #this represents the x axis

    ER_S_array_RE = getS("randomError", ER, fraction) #get the normalized S of the LCC for ER failure case
    SF_S_array_RE = getS("randomError", SF, fraction) #get the normalized S of the LCC for SF failure case
    
    ER_S_array_TA = getS("targetAttack", ER, fraction) #get the normalized S of the LCC for ER attack case
    SF_S_array_TA = getS("targetAttack", SF, fraction) #get the normalized S of the LCC for SF attack case

    plt.figure(figsize=(7, 5))
    plt.title("Failure and Attack Scenarios")
    plt.xlabel("f (fraction of removed nodes)")
    plt.ylabel("S (normalized node size) of LCC (largest connected component)")
    plt.plot(fraction, ER_S_array_RE, color = 'b', marker='o', label= 'ER Failure') #x-axis is fraction of nodes to be removed, y-axis is normalized S of the LCC for ER failure case 
    plt.plot(fraction, SF_S_array_RE, color = 'r', marker='s', label= 'SF Failure') #x-axis is fraction of nodes to be removed, y-axis is normalized S of the LCC for SF failure case 
    plt.plot(fraction, ER_S_array_TA, color = 'b', marker='>', label= 'ER Attack') #x-axis is fraction of nodes to be removed, y-axis is normalized S of the LCC for ER attack case 
    plt.plot(fraction, SF_S_array_TA, color = 'r', marker='^', label= 'SF Attack') #x-axis is fraction of nodes to be removed, y-axis is normalized S of the LCC for SF attack case 
    plt.legend(loc='upper right')
    plt.show()
```
